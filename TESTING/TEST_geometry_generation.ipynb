```python
import pandas as pd
import numpy as np
import csv

def generate_random_dataset(num_samples=1):
    all_data = []

    # Definieer de vaste topologie (edges blijven hetzelfde voor deze structuur)
    # v0-v3: base corners, v4: apex
    edges_map = {
        'v0': 'e0 e3 e4',
        'v1': 'e0 e1 e5',
        'v2': 'e1 e2 e6',
        'v3': 'e2 e3 e7',
        'v4': 'e4 e5 e6 e7'
    }

    print(f"Genereren van {num_samples} samples met random geometry...")

    for i in range(num_samples):
        # 1. Random Parameters
        # Simuleer variatie in houtafmetingen (bijv. planken/balken van verschillende formaten)
        width = np.random.uniform(0.5, 1.5)   # Random breedte
        length = np.random.uniform(0.5, 1.5)  # Random lengte
        height = np.random.uniform(0.5, 2.0)  # Random hoogte

        # 2. Noise Injection (belangrijk voor reclaimed timber simulatie)
        # Voegt kleine afwijkingen toe aan elke vertex zodat het niet perfect steriel is.
        noise_level = 0.02

        # Functie om coördinaat met noise te genereren
        def get_coord(x, y, z):
            nx = x + np.random.normal(0, noise_level)
            ny = y + np.random.normal(0, noise_level)
            nz = z + np.random.normal(0, noise_level)
            return (f"{nx:.3f},{ny:.3f},{nz:.3f}")

        # 3. Bereken Vertex Posities
        # Base op z=height, Apex op z=0 (zoals in je voorbeeld, omgekeerde piramide)
        # v0: (0, 0, h)
        # v1: (w, 0, h)
        # v2: (w, l, h)
        # v3: (0, l, h)
        # v4: (w/2, l/2, 0) -> Center point

        coords = {
            'v0': get_coord(0, 0, height),
            'v1': get_coord(width, 0, height),
            'v2': get_coord(width, length, height),
            'v3': get_coord(0, length, height),
            'v4': get_coord(width/2, length/2, 0)
        }

        # 4. Data toevoegen aan lijst
        for v_name in ['v0', 'v1', 'v2', 'v3', 'v4']:
            all_data.append({
                'sample_id': i,              # Unieke ID voor elke sample
                'Vertex': v_name,
                'Coordinates': coords[v_name],
                'Edges': edges_map[v_name]
            })

    # Maak DataFrame
    df = pd.DataFrame(all_data)

    filename = f"TEST_geometry_dataset_{amount_samples}.csv"
    df.to_csv(filename, index=False, sep=';')

    print(f"Klaar! Dataset geëxporteerd naar '{filename}'")

    # Gebruik print() in plaats van alleen 'df' om de Colab error te voorkomen
    print("\nPreview van hoe de data er in de CSV uitziet (eerste 5 regels):")
    # We lezen het bestand als tekst om te bewijzen dat er geen quotes zijn
    with open(filename, 'r') as f:
        for _ in range(5):
            print(f.readline().strip())

    return df

# Genereer de dataset
amount_samples = 10000
df_dataset = generate_random_dataset(amount_samples)
```
