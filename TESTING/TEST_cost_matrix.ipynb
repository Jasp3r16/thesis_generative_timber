```python
def calculate_lca_impact(stock, distance_km, process_penalty_kg=0):
    """
    Berekent de totale kg CO2eq voor een balk.
    """
    # 1. Productie Impact (A1-A3)
    # Reclaimed = 20, Nieuw = 450 (voorbeeldwaarden)
    gwp_factor = 450 if stock['is_new'] else 20
    production_impact = stock['volume_m3'] * gwp_factor

    # 2. Transport Impact (A4)
    # Gewicht in tonnen (density in kg/m3 / 1000)
    weight_tons = (stock['volume_m3'] * stock['density']) / 1000

    # Transport factor: 0.10 voor groot transport (nieuw), 0.20 voor klein (reclaimed)
    transport_factor = 0.10 if stock['is_new'] else 0.20
    transport_impact = weight_tons * distance_km * transport_factor

    # 3. Processing (C3/A3)
    # Voeg boete toe als er spijkers zijn (bijv. 2.5 kg CO2)
    processing_impact = process_penalty_kg if stock['has_nails'] else 0

    total_co2 = production_impact + transport_impact + processing_impact
    return total_co2
```
