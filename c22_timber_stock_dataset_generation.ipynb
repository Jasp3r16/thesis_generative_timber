### Timber stock dataset generation script
**Goal:** Generate the timber stock datasets that the algorithm can use to choose elements for the design.

**Inputs:**
*   Parameters for stock properties

**Outputs:**
*   CSV file with datasets for new stock and reclaimed stock

# STOCK GENERATION

```python
# ==========================================
# GENERATE NEW STOCK
# ==========================================
import pandas as pd
import config
from generation_timber import generate_new_timber_catalog, generate_reclaimed_stock

df_new_stock = generate_new_timber_catalog()
display(df_new_stock.head(3))

# ==========================================
# GENERATE RECLAIMED STOCK
# ==========================================
df_reclaimed_stock = generate_reclaimed_stock()
display(df_reclaimed_stock.head(3))
```

```python
# ==========================================
# TABLE WITH DEPTH AND WIDTH
# ==========================================
print("\n" + "="*50)
print("📊 TABLE: ALL DEPTH AND WIDTH COMBINATIONS")
print("="*50)

depth_width_combinations = df_new_stock[['Depth', 'Width']].drop_duplicates().sort_values(by=['Depth', 'Width']).reset_index(drop=True)

print(f"\n{'Basic size':<15} {'D (mm)':<12} {'W (mm)':<12}")
print("-" * 40)

for idx, (_, row) in enumerate(depth_width_combinations.iterrows(), 1):
    depth = int(row['Depth'])
    width = int(row['Width'])
    print(f"{idx:<15} {depth:<12} {width:<12}")

print("-" * 40)
print(f"Total number of combinations: {len(depth_width_combinations)}")

# depth_width_combinations.to_csv('depth_width_combinations.csv', index=False)
# print("\n✅ Exported to 'depth_width_combinations.csv'")
```

# EXPORTING

```python
import config
# Combine both datasets into one large matrix
df_combined_stock = pd.concat([df_new_stock, df_reclaimed_stock], ignore_index=True)

# Show a random selection from the combined dataset
print("\nPreview of the combined inventory (mix of NS and RS):")
display(df_combined_stock.sample(6))

# Export
dataset_filename = 'complete_timber.csv'
df_combined_stock.to_csv(config.TIMBER_STOCK_PATH / dataset_filename, index=False, sep=';')
# df_combined_stock.to_csv(dataset_filename, index=False, sep=';')

print(f"\nEverything was successfully combined and saved as '{dataset_filename}'.")
```
