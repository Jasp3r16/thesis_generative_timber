```python
import pandas as pd
import numpy as np
import csv

def generate_frame_dataset(num_samples=1000):
    all_data = []

    # --- 1. Topologie Definitie ---
    # We hebben nu 8 punten (4 vloer, 4 dak).
    # v0-v3 = Voet van de kolommen (vast op de grond)
    # v4-v7 = Top van de kolommen (variabele hoogte)

    # Edge Mapping: Welke balken komen samen in welk punt?
    # e0-e3 zijn de kolommen (verticaal)
    # e4-e7 zijn de liggers (horizontaal/schuin)
    edges_map = {
        # Voetpunten (alleen verbonden met de kolom omhoog)
        'v0': 'e0',          # Kolom 1
        'v1': 'e1',          # Kolom 2
        'v2': 'e2',          # Kolom 3
        'v3': 'e3',          # Kolom 4

        # Toppunten (verbonden met kolom omlaag + 2 liggers)
        'v4': 'e0 e4 e7',    # Bovenkant Kolom 1 + Ligger naar 2 + Ligger naar 4
        'v5': 'e1 e4 e5',    # Bovenkant Kolom 2 + Ligger naar 1 + Ligger naar 3
        'v6': 'e2 e5 e6',    # Bovenkant Kolom 3 + Ligger naar 2 + Ligger naar 4
        'v7': 'e3 e6 e7'     # Bovenkant Kolom 4 + Ligger naar 3 + Ligger naar 1
    }

    print(f"Genereren van {num_samples} frames met variabele kolomhoogtes...")

    for i in range(num_samples):
        # --- 2. Variabelen Genereren ---
        # De standaard hoogte is 3 meter.
        # We genereren 4 onafhankelijke afwijkingen tussen -1 en +1 meter.
        h_base = 3.0

        h_cols = [
            h_base + np.random.uniform(-1, 1), # Hoogte kolom 1 (v4)
            h_base + np.random.uniform(-1, 1), # Hoogte kolom 2 (v5)
            h_base + np.random.uniform(-1, 1), # Hoogte kolom 3 (v6)
            h_base + np.random.uniform(-1, 1)  # Hoogte kolom 4 (v7)
        ]

        # Vaste afmeting van het vierkant (3x3 meter)
        L = 3.0

        # Helper voor string formatting (3 decimalen)
        def fmt(x, y, z):
            return f"{x:.3f},{y:.3f},{z:.3f}"

        # --- 3. Coördinaten Berekenen ---
        coords = {
            # --- Vloer punten (z=0) ---
            'v0': fmt(0, 0, 0),
            'v1': fmt(L, 0, 0),
            'v2': fmt(L, L, 0),
            'v3': fmt(0, L, 0),

            # --- Dak punten (z=variabel) ---
            # Ze staan recht boven de vloerpunten, maar met de random hoogtes
            'v4': fmt(0, 0, h_cols[0]),
            'v5': fmt(L, 0, h_cols[1]),
            'v6': fmt(L, L, h_cols[2]),
            'v7': fmt(0, L, h_cols[3])
        }

        # --- 4. Data Opslaan ---
        for v_name in ['v0', 'v1', 'v2', 'v3', 'v4', 'v5', 'v6', 'v7']:
            all_data.append({
                'sample_id': i,
                'Vertex': v_name,
                'Coordinates': coords[v_name],
                'Edges': edges_map[v_name]
            })

    df = pd.DataFrame(all_data)

    # --- 5. Export ---
    # Semicolon separator om quotes te voorkomen
    filename = 'structure_frames_variable_height.csv'
    df.to_csv(filename, index=False, sep=';')

    print(f"Dataset '{filename}' gegenereerd.")
    return df

# Uitvoeren
df = generate_frame_dataset(10000)

# Preview tonen
print("\nPreview van data (eerste 8 regels = 1 volledig frame):")
print(df.head(8))
```
